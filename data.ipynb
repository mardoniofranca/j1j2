```python
import pandas as pd
import os
```

```python
def count_files(path):
    return sum([len(files) for _, _, files in os.walk(path)])
```

```python
n = count_files('data/')
n 
```

```python
#teste para 1 linha
df = pd.DataFrame()
i = 0
S_N = 'data/df_' +  str(i) + '.csv'
df_ler = pd.read_csv(S_N)
df = pd.concat([df,df_ler]) 
```

```python
df.head()
```

```python
#for para todas as linhas
s = count_files('data/')
df = pd.DataFrame()
for i in range(0,s):
    S_N = 'data/df_' +  str(i) + '.csv'
    #ler a linha
    df_ler = pd.read_csv(S_N)
    df = pd.concat([df,df_ler]) 
```
